```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
import sklearn
import seaborn as sns
from scipy import stats
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
digits.data
```

```python
plt.gray()
plt.matshow(digits.images[17])
plt.show()
```

```python
X = digits.data
y = digits.target
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(perplexity= 30, n_iter=1000, random_state=0)
```

```python
tsne_obj= tsne.fit_transform(X)
```

```python
tsne_obj
```

```python
tsne_df = pd.DataFrame({'X':tsne_obj[:,0],
                        'Y':tsne_obj[:,1],
                        'digit':y})
tsne_df.head(10)
```

```python
sns.scatterplot(x="X", y="Y",
              hue="digit",
              palette=['purple','red','orange','brown','blue',
                       'dodgerblue','green','lightgreen','darkcyan', 'black'],
              legend='full',
              data=tsne_df);
```
